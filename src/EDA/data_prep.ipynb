```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
json_dir = Path("../../data/raw")

dir_names = []
for json_file in json_dir.glob("*.json"):
    if json_file.name not in {'scraped_videos_tracker.json'}:
        dir_names.append(json_file.name)
dir_names
```

```python
dfs = []
for dir_name in dir_names:
    df = pd.read_json(json_dir / dir_name)
    df['channel_name'] = dir_name.split('_')[0]
    dfs.append(df)
display(dfs[0].head())
display(dfs[1].head())
```

```python
df = pd.concat(dfs)
display(df.sample(10))
display(df.info())
```

```python
import re

def tokenize(text):
    text = text.lower()
    return re.findall(r"\b[a-z0-9]+\b", text)

df['tokens'] = df['text'].apply(tokenize)
df.sample(5)
```

```python
original_df = df.copy()
df = df.explode('tokens')
df = df.rename(columns={"tokens": "word"})
```

```python
display(df.sample(5))
display(original_df.sample(5))
```

```python
latest = df["published_at"].max()

df["week"] = ((latest - df["published_at"]).dt.days // 7)
df = df[df["week"] < 20]
```

```python
df.sample(5)
```

```python
agg = df.groupby(["word", "week"]).agg(
    count=("word", "size"),
    likes=("likes", "sum")
).reset_index()
```

```python
agg.sample(5)
```

```python
count_pivot = agg.pivot(index="word", columns="week", values="count").fillna(0).astype(int)
count_pivot.columns = [f"week_{c}_count" for c in count_pivot.columns]
```

```python
count_pivot.sample(5)
```

```python
likes_pivot = agg.pivot(index="word", columns="week", values="likes").fillna(0).astype(int)
likes_pivot.columns = [f"week_{c}_likes" for c in likes_pivot.columns]
```

```python
likes_pivot.sample(5)
```

```python
final_df = pd.concat([count_pivot, likes_pivot], axis=1).fillna(0)
```

```python
final_df.sample(5)
```
